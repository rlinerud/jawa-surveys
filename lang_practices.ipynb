```python
import re

def clean(s: str) -> str:
    if pd.isna(s):
        return s
    else:
        s = s.lower()
        s = re.sub('baha?sa', '', s)
        s = re.sub('b\.', '', s)
        s = re.sub('bhs', '', s)
        s = re.sub('dialek', '', s)
        s = re.sub('/', ',', s)
        s = re.sub('dan', ',', s)
        s = re.sub(', *,', ',', s)
        s = s.strip()
        s = re.sub('  +', ' ', s)
        s = re.sub(',$', '', s)
        return s

usages_unpacked = usages_unpacked.applymap(clean)
usages_unpacked
```

```python
def unpack(s: str) -> frozenset[str]:
    if pd.isna(s):
        return frozenset()
    else:
        langs = re.split(' *, *', s)
        return frozenset(langs)
```

```python
import numpy as np

def langs_in_row(context_langs: pd.DataFrame) -> pd.DataFrame:
    lang_sets = np.empty(len(context_langs), dtype=frozenset)
    for index, row in context_langs.iterrows():
        langs = set()
        for entry in row:
            langs |= entry

        lang_sets[index] = frozenset(langs)

    return pd.DataFrame(lang_sets, columns=['languages'])
```

```python
from collections import defaultdict
from more_itertools import powerset
import re

class PerceptronTrainer:
    def __init__(self, ibu_data: pd.DataFrame,
                 respondent_data: pd.DataFrame,
                 ibu_dari_ibu_data: pd.DataFrame,
                 ibu_dari_bapak_data: pd.DataFrame) -> None:
        self.ibu_dari_ibu_maps = self.context_map(ibu_dari_ibu_data)
        self.ibu_dari_bapak_maps = self.context_map(ibu_dari_bapak_data)
        self.ibu_maps = self.context_map(ibu_data)
        self.respondent_maps = self.context_map(respondent_data)
        self.weights = defaultdict(int)
        self.contexts = {header for header in respondent_data}
        self.tags = [frozenset(contexts) for contexts in powerset(self.contexts)]

    def context_map(self, data: pd.DataFrame) -> pd.DataFrame:
        '''
        Maps the language labels in each row of a DataFrame to
        their corresponding context sets
        '''
        maps = np.empty(len(data), dtype=dict)
        for i, row in data.iterrows():
            cmap = defaultdict(frozenset)
            for j, entry in enumerate(row):
                for lang in entry:
                    cmap[lang] |= {data.columns[j]}

            for lang in cmap:
                cmap[lang] = frozenset(cmap[lang])

            maps[i] = cmap

        return pd.DataFrame(maps, columns=['context_maps'])

    def viterbi(self, langs: list[str], row: int) -> tuple[int, list[frozenset[str]]]:
        '''
        Finds the optimal sequence of context sets for the given list of languages
        '''

        # Initialize the delta table
        deltas = {(tag, -1): (float('-inf'), []) for tag in self.tags}
        deltas[frozenset('START'), -1] = 0.0, []

        def get_delta(curr_tag: frozenset[str], curr_idx: int) -> tuple[int, list[frozenset[str]]]:
            '''
            Helper method for the viterbi method
            '''

            if (curr_tag, curr_idx) in deltas:
                return deltas[curr_tag, curr_idx]

            else:
                # Get all possible cumulative Viterbi scores for curr_tag and curr_idx
                candidates = []
                for prev_tag in self.tags + [frozenset('START')]:
                    prev_delta, back_pointer = get_delta(prev_tag, curr_idx - 1)
                    ibu_dari_ibu_set = self.ibu_dari_ibu_maps.iloc[row, 0][langs[curr_idx]]
                    ibu_dari_bapak_set = self.ibu_dari_bapak_maps.iloc[row, 0][langs[curr_idx]]
                    ibu_set = self.ibu_maps.iloc[row, 0][langs[curr_idx]]

                    # Calculate the score for the current tag under consideration
                    curr_score = 0
                    curr_score += self.weights['emits', curr_tag, langs[curr_idx]]
                    curr_score += self.weights['subset', curr_tag, ibu_set]
                    curr_score += self.weights['superset', curr_tag, ibu_set]
                    curr_score += self.weights['subset', curr_tag, ibu_dari_ibu_set]
                    curr_score += self.weights['superset', curr_tag, ibu_dari_ibu_set]
                    curr_score += self.weights['subset', curr_tag, ibu_dari_bapak_set]
                    curr_score += self.weights['superset', curr_tag, ibu_dari_bapak_set]
                    #curr_score += self.weights['equal', curr_tag, ibu_set]
                    for context in curr_tag:
                        curr_score += self.weights['in', context, langs[curr_idx]]

                    # Add the current score to the best previous score and save it
                    candidates.append((prev_delta + curr_score, back_pointer + [curr_tag]))

                # Store and return the maximum score found
                deltas[curr_tag, curr_idx] = max(candidates, key=lambda x: x[0])
                return deltas[curr_tag, curr_idx]

        # Fill in the delta table
        for idx in range(len(langs)):
            for tag in self.tags:
                get_delta(tag, idx)

        # Return the maximum delta value that's in the final column of the table
        return max([deltas[tag, idx] for tag, idx in deltas if idx == len(langs) - 1],
                   key=lambda x: x[0])

    def train_to(self, filename: str) -> None:
        '''
        Starts training over the data set and writes checkpoints
        to files name after the specified filename
        '''
        curr_iter, num_wrong_tags = 0, 1
        while num_wrong_tags > 0:
            num_wrong_tags, num_wrong_sequences = 0, 0
            print(f'Working on iteration {curr_iter}...')

            for i, row in self.respondent_maps.iterrows():
                ibu_cmap = self.ibu_maps.iloc[i, 0]
                ibu_dari_ibu_cmap = self.ibu_dari_ibu_maps.iloc[i, 0]
                ibu_dari_bapak_cmap = self.ibu_dari_bapak_maps.iloc[i, 0]
                respondent_cmap = row[0]
                for lang in respondent_cmap:
                    ibu_cmap[lang] |= frozenset()

                for lang in ibu_cmap:
                    ibu_dari_ibu_cmap[lang] |= frozenset()
                    
                for lang in ibu_dari_ibu_cmap:
                    ibu_dari_bapak_cmap[lang] |= frozenset()

                cmap_entries = [(lang, respondent_cmap[lang]) for lang in ibu_dari_bapak_cmap]
                langs = [entry[0] for entry in cmap_entries]
                gold_tags = [entry[1] for entry in cmap_entries]

                _, estimated_tags = self.viterbi(langs, i)
                if gold_tags != estimated_tags:
                    num_wrong_sequences += 1
                    gold_tags = [frozenset('START')] + gold_tags
                    estimated_tags = [frozenset('START')] + estimated_tags
                    for j in range(len(langs)):
                        ibu_set = self.ibu_maps.iloc[i, 0][langs[j]]
                        ibu_dari_ibu_set = self.ibu_dari_ibu_maps.iloc[i, 0][langs[j]]
                        ibu_dari_bapak_set = self.ibu_dari_bapak_maps.iloc[i, 0][langs[j]]

                        # Emissions
                        self.weights['emits', gold_tags[j + 1], langs[j]] += 1
                        self.weights['emits', estimated_tags[j + 1], langs[j]] -= 1

                        # ibu Subsets
                        if gold_tags[j + 1] < ibu_set:
                            self.weights['subset', gold_tags[j + 1], ibu_set] += 1

                        if estimated_tags[j + 1] < ibu_set:
                            self.weights['subset', estimated_tags[j + 1], ibu_set] -= 1

                        # ibu Supersets
                        if gold_tags[j + 1] > ibu_set:
                            self.weights['superset', gold_tags[j + 1], ibu_set] += 1

                        if estimated_tags[j + 1] > ibu_set:
                            self.weights['superset', estimated_tags[j + 1], ibu_set] -= 1

                        # ibu dari ibu subsets
                        if gold_tags[j + 1] < ibu_dari_ibu_set:
                            self.weights['subset', gold_tags[j + 1], ibu_dari_ibu_set] += 1

                        if estimated_tags[j + 1] < ibu_dari_ibu_set:
                            self.weights['subset', estimated_tags[j + 1], ibu_dari_ibu_set] -= 1

                        # ibu dari ibu supersets
                        if gold_tags[j + 1] > ibu_dari_ibu_set:
                            self.weights['superset', gold_tags[j + 1], ibu_dari_ibu_set] += 1

                        if estimated_tags[j + 1] > ibu_dari_ibu_set:
                            self.weights['superset', estimated_tags[j + 1], ibu_dari_ibu_set] -= 1
                            
                        # ibu dari bapak subsets
                        if gold_tags[j + 1] < ibu_dari_bapak_set:
                            self.weights['subset', gold_tags[j + 1], ibu_dari_bapak_set] += 1

                        if estimated_tags[j + 1] < ibu_dari_bapak_set:
                            self.weights['subset', estimated_tags[j + 1], ibu_dari_bapak_set] -= 1

                        # ibu dari bapak supersets
                        if gold_tags[j + 1] > ibu_dari_bapak_set:
                            self.weights['superset', gold_tags[j + 1], ibu_dari_bapak_set] += 1

                        if estimated_tags[j + 1] > ibu_dari_bapak_set:
                            self.weights['superset', estimated_tags[j + 1], ibu_dari_bapak_set] -= 1

                        # Equals
                        if gold_tags[j + 1] == ibu_set:
                            self.weights['equals', gold_tags[j + 1], ibu_set] += 1

                        if estimated_tags[j + 1] == ibu_set:
                            self.weights['equals', estimated_tags[j + 1], ibu_set] -= 1
                        
                        for context in gold_tags[j + 1]:
                            self.weights['in', context, langs[j]] += 1
                        
                        for context in estimated_tags[j + 1]:
                            self.weights['in', context, langs[j]] -= 1


                        # Count number of individual tags that are wrong
                        if gold_tags[j + 1] != estimated_tags[j + 1]:
                            num_wrong_tags += 1

            # Output a new (sorted) feature weights file
            with open(f'{filename}{curr_iter}.txt', 'w') as file:
                emissions = [feature for feature in self.weights
                            if feature[0] == 'emits' and feature[1] != frozenset('START')]
                subsets = [feature for feature in self.weights
                           if feature[0] == 'subset' and feature[1] != frozenset('START')]
                supersets = [feature for feature in self.weights
                             if feature[0] == 'superset' and feature[1] != frozenset('START')]
                #equals = [feature for feature in self.weights
                #             if feature[0] == 'equals' and feature[1] != frozenset('START')]
                ins = [feature for feature in self.weights
                             if feature[0] == 'in' and feature[1] != 'START']

                # Write valid emission features first
                for emission in sorted(emissions, key=lambda x: (self.tags.index(x[1]), x[2])):
                    file.write(f'{emission} {self.weights[emission]}\n')

                # Then subset features
                for subset in sorted(subsets, key=lambda x: (self.tags.index(x[1]), x[2])):
                    file.write(f'{subset} {self.weights[subset]}\n')

                # Then superset features
                for superset in sorted(supersets, key=lambda x: (self.tags.index(x[1]), x[2])):
                    file.write(f'{superset} {self.weights[superset]}\n')

                #for equal in sorted(equals, key=lambda x: (self.tags.index(x[1]), x[2])):
                #    file.write(f'{equal} {self.weights[equal]}\n')
                
                for in_feat in sorted(ins, key=lambda x: (x[1], x[2])):
                    file.write(f'{in_feat} {self.weights[in_feat]}\n')

            # Print minimal debug info to standard output
            plural_tags = 's' if num_wrong_tags != 1 else ''
            plural_sequences = 's' if num_wrong_sequences != 1 else ''
            print(f'Finished iteration {curr_iter} with {num_wrong_tags} incorrect tag{plural_tags}',
                  f'and {num_wrong_sequences} incorrect tag sequence{plural_sequences}')

            curr_iter += 1

    def load_weights(self, filename: str) -> None:
        '''
        Loads a set of saved weights from the specified weights file
        '''
        with open(filename) as file:
            pattern = re.compile(r'(\(.*\)) (-?\d+)')
            for line in file:
                match = re.fullmatch(pattern, line.strip())
                lang_set, score = match[1], match[2]
                self.weights[eval(lang_set)] = int(score)

    def tag_langs(self, context_langs: pd.DataFrame) -> None:
        '''
        Tags the specified DataFrame of language labels
        using the weights stored in memory
        '''
        cmap = self.context_map(context_langs)
        ratios_correct, num_tagged = [], 0
        precision_ratios, recall_ratios = [], []
        langs = langs_in_row(context_langs)
        for i, row in langs.iterrows():
            num_correct = 0
            lang_list = list(row[0])
            max_score, sequence = self.viterbi(lang_list, i)
            num_tagged += 1
            precision = 0.0
            recall = 0.0
            for idx in range(len(sequence)):
                num_relevant = 0
                for context in sequence[idx]:
                    if context in cmap.iloc[i, 0][lang_list[idx]]:
                        num_relevant += 1

                if sequence[idx]:
                    precision += num_relevant / len(sequence[idx])
                    
                if cmap.iloc[i, 0][lang_list[idx]]:
                    recall += num_relevant / len(cmap.iloc[i, 0][lang_list[idx]])

                if sequence[idx] == cmap.iloc[i, 0][lang_list[idx]]:
                    num_correct += 1
                #else:
                #    print(sequence[idx], cmap.iloc[i, 0][lang_list[idx]])

            if len(lang_list) > 0:
                precision_ratios.append(precision / len(lang_list))
                recall_ratios.append(recall / len(lang_list))
                ratios_correct.append(num_correct / len(lang_list))

        if ratios_correct:
            print(f'Recorded accuracy: {sum(ratios_correct) / len(ratios_correct)}')
                                                    
        if precision_ratios:
            print(f'Recorded precision: {sum(precision_ratios) / len(precision_ratios)}')

        if recall_ratios:
            print(f'Recorded recall: {sum(recall_ratios) / len(recall_ratios)}')
```
