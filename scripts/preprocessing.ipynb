```python
import pandas as pd
```

```python
sections1_7 = pd.read_csv('allquestionnaires_section5-7all.csv', dtype=str)
sections1_7 = sections1_7.drop(columns=[
    'Person_ID',
    'Comments_page5',
    'z_calc_Button_label_locations_list',
    'z_calc_Count_other_people',
    'kinships::z_calc_Beliau_label',
    'kinships::z_calc_Where_label',
    'kinships::z_calc_With_who_label',
])

sections1_7.columns = [
    'survey_location',
    'session',
    'birth_year',
    'home_location',
    'sex',
    'caretaker',
    'relative_ethnicity',
    'relative_type',
    'relative_curr_home_lang',
    'relative_past_home_lang',
    'relative_work_lang',
    'relative_neighbor_lang',
    'relative_partner_lang',
    'relative_family_lang',
    'relative_education',
    'relative_past_home',
    'relative_curr_home',
    'relative_profession',
    'relative_religion',
    'relative_birth_year',
    'relative_first_lang',
    'relative_other_langs'
]

sections1_7.head(20)
```

```python
section9 = pd.read_csv('allquestionnaires_section9.csv', header=None, dtype=str)
section9.columns = ['survey_location', 'session', 'id', 'birth_year', 'home_location', 'sex', 'visit_frequency']
section9 = section9.drop(columns=['id'])

section9
```

```python
sections1_9 = sections1_7.merge(section9, how='left')

cols = [
    'survey_location',
    'session',
    'birth_year',
    'home_location',
    'sex',
    'caretaker',
    'visit_frequency',
    'relative_type',
    'relative_curr_home_lang',
    'relative_past_home_lang',
    'relative_work_lang',
    'relative_neighbor_lang',
    'relative_partner_lang',
    'relative_family_lang',
    'relative_ethnicity',
    'relative_education',
    'relative_past_home',
    'relative_curr_home',
    'relative_profession',
    'relative_religion',
    'relative_birth_year',
    'relative_first_lang',
    'relative_other_langs'
]

sections1_9 = sections1_9[cols]

sections1_9.head(20)
```

```python
sections1_9.to_csv('allquestionnaires_sections1-9.csv', index=False)
```

```python
usages_unpacked = pd.read_csv('allquestionnaires_sections1-9_unpacked_usages.csv', dtype=str)
usages_unpacked
```

```python
usages_unpacked = usages_unpacked.applymap(lambda x: x if pd.isna(x) else x.lower())
usages_unpacked
```

```python
usages_unpacked.to_csv('usages_unpacked_lowered.csv', index=False)
```

```python
import re

usages_unpacked = usages_unpacked.applymap(lambda x: x if pd.isna(x) else re.sub('bahasa ', '', x))
usages_unpacked
```

```python
usages_unpacked.to_csv('usages_unpacked_lowered_trimmed.csv', index=False)
```
