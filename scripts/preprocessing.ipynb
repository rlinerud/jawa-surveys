```python
import pandas as pd

sections1_7 = pd.read_csv('allquestionnaires_section5-7all.csv', dtype=str)
sections1_7 = sections1_7.drop(columns=[
    'Person_ID',
    'Comments_page5',
    'z_calc_Button_label_locations_list',
    'z_calc_Count_other_people',
    'kinships::z_calc_Beliau_label',
    'kinships::z_calc_Where_label',
    'kinships::z_calc_With_who_label',
])

sections1_7.columns = [
    'survey_location',
    'session',
    'birth_year',
    'home_location',
    'sex',
    'caretaker',
    'relative_ethnicity',
    'relative_type',
    'relative_curr_home_lang',
    'relative_past_home_lang',
    'relative_work_lang',
    'relative_neighbor_lang',
    'relative_partner_lang',
    'relative_family_lang',
    'relative_education',
    'relative_past_home',
    'relative_curr_home',
    'relative_profession',
    'relative_religion',
    'relative_birth_year',
    'relative_first_lang',
    'relative_other_langs'
]

sections1_7.head(20)
```

```python
section9 = pd.read_csv('allquestionnaires_section9.csv', header=None, dtype=str)
section9.columns = ['survey_location', 'session', 'id', 'birth_year', 'home_location', 'sex', 'visit_frequency']
section9 = section9.drop(columns=['id'])

section9
```
